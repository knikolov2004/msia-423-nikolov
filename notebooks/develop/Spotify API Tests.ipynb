```python
def analysis_df(playlist_URI, analysis_object="sections"):
    '''
    Extracting information about song objects to a structured dataframe. 
    Input playlist URI and analysis object (default is sections).
    Analysis objects can be (bars, beats, sections, segments, tatums)
    '''
    playlist = sp.user_playlist(sp_user, playlist_URI) 
    songs = playlist["tracks"]["items"]

    ids, names, artists = (), (), ()
    for i in range(len(songs)): 
        ids += (songs[i]["track"]["id"],)
        names += (songs[i]['track']['name'],)
        artists += (songs[i]['track']['artists'][0]['name'],)
    
    full_sections = pd.DataFrame()
    for i, n, a in zip(ids, names, artists):
        analysis = sp.audio_analysis(i)
        sections = pd.DataFrame(analysis[analysis_object])
        sections['song_artist'] = a
        sections['song_name'] = n
        full_sections = pd.concat([full_sections, sections], sort = False, ignore_index=True)
    
    return full_sections
```

```python
def get_track_ids(sp_user, playlist_uri):
    """Get individual URI's for each track in a Spotify playlist

    Args:
        sp_user (`str`): Spotify user URI
        playlist_uri (`str`): Spotify playlist URI

    Returns:
        id_list: comma separated list of track URI's
    """
    playlist = sp.user_playlist(sp_user, playlist_uri)
    tracks = playlist["tracks"]["items"]
    id_list = []
    for i in tracks:
        id_list.append(i["track"]["id"])
    return id_list
```

```python
def write_records(records, file_location):
    """Persist API response set to file.

    Args:
        records (`:obj:`list` of :obj:`str`): The list of Spotify features.
        file_location (`str`): Location to write file to.

    Returns:
        None
    """

    if not file_location:
        raise FileNotFoundError

    num_records = len(records)
    print("Writing {} records to {}".format(num_records, file_location))

    with open(file_location, "w+") as output_file:
        json.dump(records, output_file, indent=2)
```
